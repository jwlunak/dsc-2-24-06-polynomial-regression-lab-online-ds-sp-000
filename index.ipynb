## Create the best plot using polynomials!

Below, we created a plot with a clearly non-linear shape.

- plot a polynomial function using `PolynomialFeatures` for polynomials up until the second, third and fourth degree.
- print out the $R^2$ value for each of the three results. Draw conclusions with respect to which degree is best.

```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline

def pol(x):
    return x * np.cos(x)

x = np.linspace(0, 12, 100)
rng = np.random.RandomState(1234)
rng.shuffle(x)
x = np.sort(x[:25])
y = pol(x) + np.random.randn(25)*2

plt.scatter(x, y, color='green', s=50, marker='.')

plt.show();
```

## Solution

```python
## your code here
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
x_plot = np.linspace(0, 12, 100)

X = x[:, np.newaxis]
X_plot = x_plot[:, np.newaxis]

poly_2 = PolynomialFeatures(2)
X_2 = poly_2.fit_transform(X)
X_2_Plot = poly_2.fit_transform(X_plot)

reg_poly_2 = LinearRegression().fit(X_2, y)
predict_x_2 = reg_poly_2.predict(X_2)

y_plot_2 = reg_poly_2.predict(X_2_Plot)

plt.plot(x_plot,y_plot_2, color = "blue")
plt.scatter(x, y, color='green', s=50, marker='.')
plt.title('Polynomial 2')
plt.show();
```

```python
# Rsquared Polynomial 2
r2_score(y, reg_poly_2.predict(X_2))
```

```python
x_plot = np.linspace(0, 12, 100)

X = x[:, np.newaxis]
X_plot = x_plot[:, np.newaxis]

poly_3 = PolynomialFeatures(3)
X_3 = poly_3.fit_transform(X)
X_3_Plot = poly_3.fit_transform(X_plot)

reg_poly_3 = LinearRegression().fit(X_3, y)
predict_x_3 = reg_poly_3.predict(X_3)

y_plot_3 = reg_poly_3.predict(X_3_Plot)

plt.plot(x_plot,y_plot_3, color = "blue")
plt.scatter(x, y, color='green', s=50, marker='.')
plt.title('Polynomial 3')
plt.show();
```

```python
# Rsquared Polynomial 3
r2_score(y, reg_poly_3.predict(X_3))
```

```python
x_plot = np.linspace(0, 12, 100)

X = x[:, np.newaxis]
X_plot = x_plot[:, np.newaxis]

poly_4 = PolynomialFeatures(4)
X_4 = poly_4.fit_transform(X)
X_4_Plot = poly_4.fit_transform(X_plot)

reg_poly_4 = LinearRegression().fit(X_4, y)
predict_x_4 = reg_poly_4.predict(X_4)

y_plot_4 = reg_poly_4.predict(X_4_Plot)

plt.plot(x_plot,y_plot_4, color = "blue")
plt.scatter(x, y, color='green', s=50, marker='.')
plt.title('Polynomial 4')
plt.show();
```
